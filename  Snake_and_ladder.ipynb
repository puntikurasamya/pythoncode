```python
import random
```

```python
print('*'*10,'SNAKE_LADDER GAME BEGINS','*'*10)
ladder={8:26,19:38,21:82,28:53,36:57,43:77,51:91,54:88,61:99,62:96,66:87}
snake={37:7,47:4,52:11,58:40,63:17,74:35,97:15}
pos1=0
pos2=0
def move(pos):
    dice=random.randint(1,6)
    print(f'dice:{dice}')
    pos+=dice
    if pos in ladder:
        print('----------hey u got a ladder--------------')
        pos=ladder[pos]
        print(f'position:{pos}')
        if pos>100:
            pos=pos-dice
            print('sorry out of the range')
            print(f'now your position:{pos}')
    elif pos in snake:
        print('---------oops you got a snake---------')
        pos=snake[pos]
        print(f'position:{pos}')
        if pos>100:
            pos=pos-dice
            print('sorry out of the range')
            print(f'now your position :{pos}')
            
    else:
        print(f'position:{pos}')
        if pos>100:
            pos=pos-dice
            print('sorry out of the range')
            print(f'now your position:{pos}')
    print('\n')
    return pos
while True:
    player1=input('Enter name of the player1 :')
    pos1=move(pos1)
    if pos1==100:
        print('='*10,'GAME OVER','='*10)
        print(' '*10,player1,' '*10)
        print('='*10,'WINNER OF THE GAME','='*10)
        break
        
    player2=input('Enter name of the player2 :')
    pos2=move(pos2)
    if pos2==100:
        print('='*10,'GAME OVER','='*10)
        print(' '*10,player2,' '*10)
        print('='*10,'WINNER OF THE GAME','='*10)
        break
    
        
```
